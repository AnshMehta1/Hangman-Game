```python
# creating visual depiction to represent number of lives left
stages_list = ['''
  +---+
  |   |
 \O/  |
  |   |
 / \  |
  =========''','''
  +---+
  |   |
 \O/  |
  |   |
      |
  =========''','''
  +---+
  |   |
  O   |
  |   |
      |
  =========''','''
  +---+
  |   |
  O   |
      |
      |
  =========''','''
  +---+
  |   |
      |
      |
      |
  ========= ''','''
  +---+
      |
      |
      |
      |
  =========  ''']
```

```python
import random as rd

# Selecting a random word from words_list :-
guess_word = rd.choice(words_list)
length = len(guess_word)

display = []

for i in range(0,length) :
    display += "_"
    
print("The number of letters in word to be guessed is "+ str(length))    
print(display)  

lives = 5

# Iterating  while loop to run the game :-
while lives > 0 :
    guess_letter = input("Guess a letter: ")
    flag = 0
    
    for i in range(0,length) :
        if guess_letter == guess_word[i] :
            flag = 1
            display[i] = guess_letter
        else :
            continue 
    
    if flag == 0 :
        print("The letter guessed is not present in word to be guessed")
        lives = lives -1
        print(display)
        print(stages_list[lives])
        print("Lives left are: " + str(lives))    
    else :
        print("The letter guessed is present in word to be guessed")
        print(display)
        print(stages_list[lives])
        print("Lives left are: " + str(lives))  
        
    if "_" not in display :
        break 

# Displaying of result :-
if lives == 0 :
    print("")
    print("!! YOU LOSE !!")
    print("")
    
    for i in range(0,length) :
        display[i] = guess_word[i]
    
    print("The correct word is: ")
    print(display)
else :
    print("")
    print("!! YOU WON !!")   
    print("")
```
